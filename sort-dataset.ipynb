```python
def ratioVideoPercent(likes,dislikes):
    likesList =[]
    for key,value in list(likes.iteritems()):
        likesList.append(value)
        
    dislikeList =[]
    for key,value in list(dislikes.iteritems()):
        dislikeList.append(value)
        
    likedislike =list(zip(likesList,dislikeList))
    ratio =[]
    for like,dislike in likedislike:
        if (like+dislike)==0:
            ratio.append(0)
        else:          
            result =(like/(like+dislike))*100
            ratio.append(result)
    print(ratio)
    return ratio
    
data["ratio"]=ratioVideoPercent(data["likes"],data["dislikes"])
    
    
```

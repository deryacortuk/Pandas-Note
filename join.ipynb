```python
import numpy as np
import pandas as pd
```

```python
set1 ={
    "a":["x1","x2","x3"],
    "b":["y1","y2","y3"],
    "c":["z1","z2","z3"]
}
set2={"a":["i1","i2","i3"],
     "b":["j1","j2","j3"],
     "c":["k1","k2","k3"]}
df1=pd.DataFrame(set1,index=[1,2,3])
df2 =pd.DataFrame(set2,index=[4,5,6])
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis=1)
```

```python
set1 ={
    "a":["a1","b2","c","a"],
    "b":["y","x","y3","b"],
    "key":["k1","k2","k3","k6"]
    
}
set2={"x":["i","x","y"],
     "y":["j","j1","j3"], 
     "key":["k5","k2","k6"]}
df1=pd.DataFrame(set1,index=[1,2,3,4])
df2 =pd.DataFrame(set2,index=[1,2,3])
```

```python
df1
```

```python
df2
```

```python
df1.merge(df2)                      #according to key
```

```python
pd.merge(df1,df2,on=["key"])
```
